```python
#Ticks, Labels, and Legends
fig= plt.figure()
ax = fig.add_subplot(111) # one column, one row, index = 1
ax.plot(randn(1000).cumsum(), label='Random Data')
```

```python
#ticks
ax.set_xticks(list(range(0,1001,250)))


#labels
ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation=30, fontsize='small')

ax.set_title('My first Matplotlip Plot with Ticks')

#labels for x and y
ax.set_xlabel('Stages')
ax.set_ylabel('Values')
ax.legend(loc='best')

fig
```

```python
#Use dictionary for title, xlabel, and ylabel
props = {'title': 'My First Matplotlib Plot with Ticks',
         'xlabel': 'Stages',
         'ylabel': 'Values'
        
        }
ax.set(**props)
fig
```
